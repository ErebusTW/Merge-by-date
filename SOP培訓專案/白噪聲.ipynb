**<font  size=6>1. 前言</font>**  
首先明確定義一點--**金融時間序列的核心特性為自相關性**；說明金融時間序列分析的核心正是挖掘該時間序列中的自相關性。一個優秀的模型應該能夠有效的刻畫原始時間序列中不同間隔的自相關性；而衡量一個模型是否適合原始時間序列的標準正是考察原始值和擬合值之間的殘差序列是否近似的為白噪聲。

白噪音正是本文的內容之一，它是時間序列分析中最基本的模型。在它的基礎上延伸出的另一個基本模型便是隨機遊走。通常，白噪音和隨機遊走被認為是用來分別描述投資品收益率和價格的最簡單模型。我們稍後會看到，對於收益率來說（特別是股指的收益率），白噪音模型並不有效。


**<font  size=6>2. 時間序列建模</font>**  

- **本質上說講，時間序列模型是一個可以「解釋」時間序列中的自相關性的數學模型。**  

能夠解釋時間序列的自相關性在量化投資領域中意義重大：

我們假設金融時間序列（例如投資品的收益率）存在未知的自相關性（當然也伴隨著噪音），而這種自相關性體現了該時間序列某種內在的特性（例如趨勢、或均值回复），而這種內在特性是可以延續的（至少在未來短時間內）。因此，我們希望透過對歷史資料的擬合找到一個合適的模型，使得它能最大程度的解釋該時間序列所表現出來的自相關性。基於未來會重複歷史的假設，我們在統計上預期這種自相關性存在於未來的序列中，由於這個模型考慮了這種自相關性，因此它將幫助我們預測未來。

時間序列分析為我們研究投資品報酬率的行為提供了有力的統計架構。  

在投資中，對收益率的預測顯然是非常有用的。如果我們能夠預測投資品的漲跌，那麼就能基於此構建一個交易策略；如果我們能夠預測收益率的波動率，那麼就可以進行風險管理（因此我們對時間序列的二階統計量——如方差——同樣感興趣）。

假設原始時間序列為${y_t}$，模型擬合出來的序列為${\rho_t}$，則殘差序列$e_t$定義為原始序列與擬合序列的差異：  
$$
e_t=y_t-\rho_t
$$  
如果模型很好的捕捉了原始時間序列的自相關性，那麼殘差序列 $e_t$ 應該近似的為白噪聲，對任何非零間隔 k ，此殘差序列的自相關係數$\rho_k$ 都應該在統計意義上不顯著的偏離0 。當然，這只是我們說這個模型是個優秀模型的充分條件，因為一個好模型最關鍵的還是能產生賺錢的交易訊號。因此，模型的檢驗最終還要看它在樣本外預測的準確性。  

時間序列建模的過程可以總結如下：



![](attachments2/f1.png)

對於一個時間序列，我們總是希望先畫出它的相關圖來看看它有什麼樣的自相關性。基於對其自相關性的認知，第二步則是選擇合適的模型，例如AR、MA 或ARMA 模型，甚至於更高階對波動率建模的GARCH 模型等。選定模型後，接下來便需要最佳化模型的參數，使其盡可能解釋時間序列的自相關性。在這一步，我們透過對殘差進行自相關性分析來判斷模型是否合適。在這方面，Ljung–Box 檢定是一個很好的方法，它同時檢驗給殘差序列各間隔的自相關係數是否顯著的不為0。在選定模型參數之後，仍需定量評估模型在樣本外預測的準確性。畢竟，對於樣本內的數據，錯誤的過度擬合總是會得到「優秀」的模型，但它們往往對樣本外數據的預測效果很差。因此，只有樣本外預測的準確度才能客觀的評價模型的好壞。如果模型的準確性較差，這表示模型有缺陷，無法充分捕捉原始序列的自相關性。這時必須考慮更換模型。這就構成了上述步驟的回饋迴路，直到最終找到一個既能解釋原時間序列自相關性，又能在樣本外有不錯的準確性的模型。之後，該模型將被用來產生交易型號並建立量化投資策略。

接下來我們來介紹一個最簡單的時間序列模型：白噪音。

**<font  size=6>3. 白噪音</font>**  

本文第一節指出，對於收益率來說，白噪音（white noise）並不是一個十分有效的模型。那為什麼我們還要研究它呢？這是因為它有一個重要的特性，即序列不相關：一個白噪聲序列中的每一個點都獨立的來自某個未知的分佈，它們滿足獨立同分佈（independent and identically distributed）。

一個（離散）白噪音的定義如下：

考慮時間序列 $w_t:t = 1, \cdots, n$。如果該序列的成分$w_t$ 滿足均值為 0 ，方差$\sigma^2$，且對於任意的$k ≥ 1$，自相關係數$\rho_k$ 均為 0 ，則稱該時間序列為一個離散的白噪音。

上面的定義並沒有假設$w_t$來自常態分佈。事實上，白噪聲對分佈沒有要求。當$w_t$ 來自常態分佈時，此序列又稱為高斯白雜訊（Gaussian white noise）。

根據白噪聲的定義，一個白噪聲序列顯然滿足平穩性要求。它的平均數和二階統計量為：

$$
\mu_w = 0  \\
$$
$$ 
\rho_k=\left\{\begin{array}{lll}
1 & \text { if } \quad k=0 \\
0 & \text { if } \quad k \neq 0
\end{array}\right.
$$

我們已經多次強調，當一個模型很好的捕捉了原始時間序列的自相關性，它的殘差序列就應該沒有任何（統計意義上顯著的）自相關性了。換句話說，一個優秀模型的殘差序列應該（近似）為一個白噪音。因此，使用白噪聲序列的性質可以幫助我們確認我們的殘差序列中沒有任何相關性了，一旦殘差序列沒有相關性便意味著模型是原始時間序列的一個良好的擬合。

在白噪音模型中，唯一的參數就是方差$\sigma^2$。這個參數可以透過歷史資料估計得到。在本系列的第一篇文章中，我們曾給出了一個白噪聲序列的相關圖（如下），該序列由標準正態分佈產生（因此為高斯白噪聲），共500個觀測值。可以看到，對圖中顯示的間隔k的取值，對所有$k ≥ 1$皆有自相關係數在統計上等於0。

**<font  size=6>4. 隨機遊走</font>**  

將白噪音模型進行一步延伸，以便得到隨機遊走（random walk）模型，它的定義如下：

對於時間序列 $x_t$，如果它滿足$x_t=x_{t-1}+w_t$，其中$w_t$是一個均值為 0 、方差為$\sigma^2$ 的白噪聲，則序列$x_t$ 為一個隨機遊走。

由定義可知，在任意 $t$時刻的$x_t$都是不超過 $t$ 時刻的所有歷史白噪音序列的總和，即：
$$
x_t=w_t+w_{t-1}+w_{t-2}+\cdots+w_{0}
$$

隨機遊走的序列平均值和變異數為：

$$
\mu_{x_t}  =0 
$$
$$
\begin{aligned}
\operatorname{Var}\left(x_t\right) & =\operatorname{Var}\left(w_t\right)+\operatorname{Var}\left(w_{t-1}\right)+\cdots+\operatorname{Var}\left(w_0\right) \\
& =t \times \operatorname{Var}\left(w_t\right) \\
& =t \sigma^2
\end{aligned}
$$
雖然均值不隨時間 $t$ 改變，但是由於變異數是 $t$ 的函數，因此隨機遊走不滿足平穩性。隨著$t$的增加， $x_t$ 的變異數增大，顯示其波動性不斷增加。對於任意給定的$k$，透過以下推導給得出隨機遊走的自協方差：  


$$
\begin{aligned}
\operatorname{Cov}\left(x_t, x_{t+k}\right) & =\operatorname{Cov}\left(x_t, x_t+w_{t+1}+\cdots+w_k\right) \\
& =\operatorname{Cov}\left(x_t, x_t\right)+\sum_{i=t+1}^k \operatorname{Cov}\left(x_t, w_i\right) \\
& =\operatorname{Cov}\left(x_t, x_t\right)+0 \\
& =t \sigma^2
\end{aligned}
$$  

上述推導中使用了獨立隨機變數的方差可加性。 有了自協方差和方差，便可以方便的求出隨機遊走的自相關函數：  

\begin{aligned}
\rho_k(t) & =\frac{\operatorname{Cov}\left(x_t, x_{t+k}\right)}{\sqrt{\operatorname{Var}\left(x_t\right)} \sqrt{\operatorname{Var}\left(x_{t+k}\right)}} \\
& =\frac{t \sigma^2}{\sqrt{t \sigma^2} \sqrt{(t+k) \sigma^2}} \\
& =\frac{1}{\sqrt{1+k / t}}
\end{aligned}  

顯然，自相關係數既是時間 $t$ 又是間隔 $k$ 的函數。$\rho$ 的表達式說明，對於一個足夠長的隨機遊走時間序列（ $t$ 很大），當考察的自相關間隔 $k$ 很小時，自相關係數近似為 1 。這是隨機遊走的一個非常重要的特性，不熟悉它往往容易造成不必要的錯誤。

舉個例子。我們通常假設股價的對數報酬率符合常態分佈，因此股價對數是一個布朗運動（隨機遊走的特殊形式）。如果目前的（對數）股價是$x_t$，由隨機遊走的特性可知， $t+1$時刻的股價的條件期望為${E}[x_{t+1} | x_t] = x_t$，即我們對下一時點的股價的最好的猜測就是當前的價格。隨機遊走是一個鞅（martingale）。

假如我們有一個預測股價的模型，而這個模型就是用 $t$ 時刻的股價作為對$t+1$時刻的股價的預測，則該模型的預測值和實際值之間的相關係數就等於股價序列的間隔為 1 的自相關係數。如果股價近似的為隨機遊走，那麼由它的性質可知，間隔為1的自相關係數非常接近 1。因此我們的股價預測模型——以今天的價格作為明天的價格的預測——的預測值和實際值之間的相關係數也非常接近 1。這會給我們造成錯覺：這個模型相當準確。不幸的是，這個模型猜測的收益率在任何時刻都為 0 ，因此它對於我們建立交易訊號毫無作用。

我看到無數的學術論文（大多是碩士論文）中，針對投資品價格本身建立自迴歸模型。獨立變數就包括歷史價格，用它們和其他一些基本面或宏觀經濟數據來預測下一個交易日的股價。從上面的分析可知，這樣的模型將會“精準的毫無用處”，因為回歸模型中歷史價格的係數之和將會非常接近 1。

任何價格序列的自回歸模型都是耍流氓。

**<font  size=6>5. 用白雜訊對收益率建模</font>**  

如果股票的對數收益率為白雜訊，那麼它的自相關係數應該在任何非零的間隔上都在統計意義上等於零。 下面我們就來看看真實的股票收益率是否滿足這一點。 為此，考慮一支個股和一個股指。

以日頻為例，通過交易日的復盤后收盤價可以算出對數收益率：  
$$
r_t=\ln x_t-\ln x_{t-1}
$$  

首先來看看個股，當考察期為過去$10$年時，萬科的對數收益率的相關圖為

![](attachments2/f6.png)

上圖指出，在間隔為2和4時，此收益率序列表現出了統計意義上顯著的相關性。當然，由於圖中的藍色區域只是95\%的置信區間，因此僅根據隨機性也很可能出現在一個或兩個間隔上的自相關係數處於置信區間之外的情況。因此，根據上面的結果，我們並不能一定就說白噪音不是個股收益率的一個適當的模型。
