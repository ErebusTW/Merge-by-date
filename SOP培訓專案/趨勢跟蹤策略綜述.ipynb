**<font  size=6> 趨勢跟蹤策略綜述 </font>**

**<font  size=5> 1.存在和邏輯 </font>**  

**1.1 無處不在**  
常常聽到別人講，趨勢是你的朋友。趨勢追蹤（trend following）可能是最容易理解的投資策略。當資產價格上漲時，它處於上行趨勢中，可考慮買入看多；當資產價格下跌時，它處於下行趨勢中，可考慮賣出看空。因此，趨勢追蹤本質上是一個時序動量（time-series momentum），它只關注資產本身的歷史走勢，而不用考慮當前時點其他資產的表現。與之相對應的另一個概念是（橫）截面動量（cross-sectional momentum），比較經典的研究來自於Jegadeesh and Titman (1993)。在某一個時刻
$
t$，比較所有資產過去一段時間的表現，做多表現相對強勢的一籃子資產（winner），做空表現相對弱勢的一籃子資產（loser）。時間序列動量是一個絕對概念，在某一時刻只和自己的歷史相比；截面動量是一個相對概念，在同一時刻進行橫向比較。

當人們談到趨勢追蹤時往往指的是時序動量，本文接下來將圍繞時序動量展開介紹。在下文的闡述中，我們會交替使用趨勢追蹤和時序動量兩種表述[1]。對於截面動量感興趣的小夥伴，請參考本網站資產定價頁面中動量因子標籤下的相關文章。

作為一種量化策略，趨勢追蹤無所不在。Moskowitz, Ooi and Pedersen (2012) 利用商品期貨、國債期貨、匯率期貨和股指期貨等58 個資產，建構了時間序列動量組合。結果發現，在1 至12 個月範圍內，收益率具有持續性，即過去上漲的資產未來上漲的機率較大，過去下跌的資產未來下跌的可能性較大；從1985 年到2009 年，時間序列動量組合表現優秀，超額報酬顯著大於0。

Hurst, Ooi and Pedersen (2017) 將研究範圍和時間區間進一步拓展至包括29 個商品、11 個權益指數、15 個國債指數和12 個匯率對，且將實證數據從1880 年開始（如果沒有期貨就用現貨拼接）。在構造趨勢追蹤組合時，同時考慮了1 個月、3 個月和12 個月三個不同頻率的週期。從1880 年到2013 年，趨勢追蹤組合年化收益14.9%，即使考慮了2% 的管理費和20% 的業績提成，仍然能獲得11.2% 的年化收益率，而組合年化波動只有9.7% ，夏普比率高達0.77。Babu et al. (2020) 在Moskowitz, Ooi and Pedersen (2012) 所涵蓋的58 個傳統資產的基礎上，增加了82 個另類資產和16 個因子收益率序列，結果顯示趨勢跟踪廣泛存在於這些資產中，並且長期有效；透過拓展新的資產類型，可顯著提高夏普比率。

Moskowitz, Ooi and Pedersen (2012)、Hurst, Ooi and Pedersen (2017) 和Babu et al. (2020) 等的研究均表明，無論從時間維度上，還是資產類別上，抑或是投資週期上，均存在趨勢追蹤策略的身影，表現十分亮眼。  

**1.2 背後邏輯**  
趨勢追蹤能夠長期保持活力（劇透！本文第9 節討論了近年來趨勢追蹤變差的原因），背後肯定有堅實的邏輯基礎。站在當前時點，一切歷史都是樣本內，因此在評估一個投資策略是否站得住腳時，如果過於關注收益率等風險收益指標，而不注重為什麼有效，再好看的數據都是脆弱的。

學術上在解釋因子的收益來源時，往往從兩個角度著手：風險補償和錯誤定價。前者是經典金融學的角度，認為一個異像如果能獲得超額收益是因為承擔了某種風險，超額收益是這種風險的補償；後者從行為金融學的角度，認為要么是某種限制阻礙了異象的消失，要嘛是投資人認知偏差導致的行為不理性。

Moskowitz, Ooi and Pedersen (2012) 試圖從風險的角度研究趨勢追蹤的收益來源，時間序列回歸結果表明趨勢追蹤並不能被現有的風險因子（如HML、SMB 等）所解釋；不僅難以確定風險來源，該文發現在市場表現很差的時候，趨勢追蹤反而表現出色，例如2008 年金融危機時CTA 基金一枝獨秀。因此，難以從風險補償的角度來理解趨勢追蹤。從行為金融的角度討論就要容易得多，Clare et al. (2016) 以及Hurst, Ooi and Pedersen (2017) 對此有過詳細討論。有效市場假說認為，當某個重要的新資訊出現時，價格會立即得到反應，以配合最新的內在價值。然而，由於投資人認知偏差的純在，市場並不是完全有效，而是慢慢吸收新資訊的影響。

大體來說，趨勢的形成可以分為三個階段，如圖1 所示。  

![](attachments/f1-1.png)

初始階段，由於錨定效應( anchoring )、處置效應( disposition effect）和非營利操作等，導致價格對資訊反應不足。

- 錨定效應：投資人往往會把自己的觀點錨定在最近歷史數據上，不願意很快改變自己的想法，當新消息到來時反應很遲鈍，造成價格反應緩慢；
- 處置效應：投資人往往會過早低賣出獲利股票以兌現收益；相反，對那些虧損的股票遲遲不肯止損。這樣的後果是，上漲不會一蹴而就，下跌不會一跌到底，造成價格慢慢的移動；
- 非營利操作：一些非營利操作，也會減緩價格的反應速度，例如中央銀行在外匯市場和固定收益市場進行斡旋，以減少匯率和利率的波動。  

一旦趨勢起來，就進入了第二階段：趨勢持續甚至過度反應，這往往是由於羊群效應（herding），以及確認和代表偏誤（confirmation bias 和representativeness）等因素導致的。

- 羊群效應：羊群效應也叫從眾效應，當投資人觀點和其他大多數投資人觀點不一致時，容易懷疑和改變自己的觀點，以使得和群體一致。在價格上漲或下跌開啟後，投資會想羊群一樣，加入到趨勢行情中來；
- 確認與代表性偏誤：行情開啟後，投資人會用最近的價格變動來推斷未來，選擇性地關注最近獲利的方向；不僅如此，一旦確認了自己的觀點，便會找各種數據和資訊支持自己的看法，這使得價格趨勢得以延續。  

最後，趨勢不會永不眠，價格不會一直朝一個方向走下去。在趨勢的末端，價格可能已經過度反應導致嚴重偏離基本面，因此最終會出現反轉趨勢宣告結束。

**<font  size=5> 2. 風險收益特徵 </font>**  
**2.1 Divergent Risk Taking**  
Jusselin et al. (2017) 指出，時序動量策略的收益特徵是凸函數（圖2）。當趨勢很小的時候，該策略的收益為負；只有當趨勢很大的時候（既可以是上漲趨勢也可以是下跌趨勢），該策略的收益為正。
![](attachments/f2.png)  
由此曲線可知，時序動量策略的虧損有限，收益無上限。值得說明的是，這裡的「虧損有限」是有條件的，它要求動量策略不能使用太高的槓桿率（當槓桿過高時，它對策略的收益和虧損的放大不是線性的），以及資產價格應該是連續的。從趨勢追蹤策略的PnL 曲線來看，它在更多的時候是連續經歷小幅回撤，然後換來一大波（大趨勢）的上漲，從而將之前的眾多小幅回撤抹平，使得該策略的長期收益為正。這個「虧損有限、收益無限」的特徵造就了趨勢策略收益分佈的正偏。這個分佈又被稱為聖杯分佈（Holy Grail distribution）。趨勢追蹤策略雖然不是聖杯，但它的收益分佈滿足聖杯分佈。
![](attachments/f3.png)  

正是因為這種獨特的特性，使得趨勢策略能在其他資產發生尾部風險的時候脫穎而出，這也就是它為什麼享有危機阿爾法的美名（見本文7.1 節）。為什麼不同策略的報酬率會表現出不同的分佈特徵？這可以從兩類不同的風險來解釋。任何投資策略想要賺錢，都需要承擔一定的風險。而策略承擔什麼類型的風險就決定了它的收益分佈具備何種的特徵。風險可以被分為convergent risks和divergent risks兩大類，因此策略也可以被視為convergent risk taking 和divergent risk taking 兩大類。
1. Convergent risks 是那些被人們理解、可透過建模來測量的風險；
2. Divergent risks 是哪些未知的風險、無法預測的風險、不能很好的被衡量的風險。  

當建構convergent risk taking 策略的時候，人們根據金融學或經濟學原理對資產的走勢有一個先驗信仰，並以此為判斷做出投資決策。另一方面，在建構divergent risk taking 策略時，人們假設資產未來的走勢是未知的、不使用任何先驗信仰來輔助判斷資產會漲還是會跌，而是根據資產價格自身的發展做出相應的判斷。舉例來說，股票市場中的價值投資正是convergent risk taking 的代表。這類策略透過深度產業研究挖掘上市公司的內在價值，當價格低於價值時則買入，耐心的等待其價格向價值回歸、無視短期的波動甚至是持續下跌，這是這類策略主動承擔的風險。而常見於CTA 的趨勢追蹤策略則是divergent risk taking 的代表。這類策略對資產的內在價值不做任何判斷，而是嚴格依據價格的走勢進行交易：價格上漲了做多、價格下跌時做空；當盈利時會逐漸加倉，當虧損時會及時清倉。正因如此，趨勢追蹤策略屬於divergent risk taking 策略，因而其收益滿足聖杯分佈的特性，這也正是趨勢追蹤策略能夠長盛不衰的原因。




**2.2 趨勢vs 波動**  
假設資產收益率滿足幾何布朗運動：
$$
\frac{dP}{P} = \mu dt + \sigma dB
$$  

其中$P$是資產價格，$μ$是漂移率，$σ$是波動率，$B$是標準布朗運動。由此模型易知（做多的）趨勢追蹤策略喜歡收益率中正的漂移率項（即$μ$越大越好），而不喜歡波動（即$σ$越小越好）。$μ$越大且/或$σ$越小說明趨勢越顯著（當然對於能做空的情況，那麼只需要$μ$的絕對值越大越好）。因此，在評估一個資產是否適合動量策略時，必須同時考慮$μ$和$σ$，所以業界認為資產的夏普率是一個評價它趨勢強弱的優秀指標。同樣，對於能做空的情況，我們會更關注夏普率的絕對值的值。  
借用選擇權定價的術語，令$S$表示時序動量策略的淨值，$P$和$σ$分別為標的資產的價格和波動率，則時序動量策略的Delta 和Vega 為：  
$$
Delta = \frac{\partial S}{\partial P}, Vege = \frac{\partial S}{\partial \sigma}
$$  
從業務上來解釋，Delta 可以理解為資產價格的變動對時序動量策略淨值的影響。以追逐上漲趨勢的做多策略為例，它應該具備正的Delta，也就是隨著價格的上漲，策略淨值上漲。另一方面，Vega 代表收益率波動對時序動量策略淨值的影響。Roncalli (2017) 指出，時序動量策略具有負的Vega，資產的短期波動會造成動量策略淨值的下降。由於“波動常有、而趨勢不常有”，因此資產的反覆波動會主宰動量策略的表現（陰跌、虧損），直到一個大趨勢到來並被策略捕捉到。以美股市場為例，由於它反映了經濟的發展和上市公司的獲利成長，市場呈現出慢牛的結構。在這種結構下，基於長週期的時序動量策略比基於短週期的時序動量策略效果要好。


**<font  size=5> 3. 趨勢的時間尺度 </font>**  
想要透過趨勢追蹤策略賺錢，最重要的是什麼？很多文獻告訴我們，趨勢追蹤中的重中之重包括trend measure（即選擇最優的指標來計算趨勢）以及position sizing（倉位管理）。這兩點對於趨勢策略至關重要。本節介紹前者，下一節將討論後者。

舉個例子，Winton (2013) 以下列來自四大類的20 個資產進行了趨勢策略時間尺度的研究。

為了進一步分析這個問題，考慮到資產收益率序列通常包含低頻均值、（非常微弱的）自相關性以及白噪聲三個部分構成，利用以下步驟人工合成資產對數收益率：  
$$
Step1:b_0 ∼ N(0,\sigma)\\
$$
$$
Step2:e_t ∼ N(0,\sigma)\\
$$
$$
Step3:b_t = \rho \times b_{t-1} + e_t  \\
$$
$$
Step4:r_t ∼ b_t-\mu_b+\mu
$$
上述過程中，$e_t$是白噪聲序列，它滿足平均值為0，標準差為$σ$的常態分佈。$r_t$是對數收益率序列，構造過程中使用了輔助序列$b_t$。首先利用迭代構造出滿足自相關性$ρ$的序列$b_t$。  
由於在建構$b_t$時可能會引入意外的平均值$μ_b$，因此在利用$b_t$ 構造$r_t$ 時，將$μ_b$ 去除並把給定的平均值$μ$ 加回來。這樣最終產生的對數報酬率序列$r_t$  同時滿足的長期平均值 $μ$  
和自相關性$ρ$。這就是上面四個式子所表達的過程。
滿足長期平均值$\mu$和自相關性$\rho$。利用這個數學模型，以$\mu$、$\rho$、$\sigma$作為輸入，透過大量模擬分析可以得到以下結論：

1. 長期均值$μ$主宰自相關性$ρ$ —— 比起縹緲的自相關性，低頻分量$μ$才是真正的趨勢（想想股市的大牛市）；

2. 收益率均值$μ$和收益率標準差$σ$的數量級之差會影響趨勢效果— 這意味著合理選擇計算趨勢的時間尺度至關重要。  

通常來說，對於日頻數據，收益率標準差比平均值高兩個數量級，在這個層級上趨勢策略難有作為。如果把時間拉長，在更低的頻率（即更長的時間尺度）上計算收益率，可以有效提高平均值並降低其標準差，當平均值和標準差相差一個數量級時，趨勢策略或大有可為。這就是為什麼同一組趨勢參數在台股好使（台股大牛市的時候日收益率均值太高）卻在美股上不好使。  

**<font  size=5> 4. 部位管理 </font>**  
倉位管理（position sizing）也被視為槓桿管理。期貨策略往往給人一種高風險高收益的印象，這種想法隱含了（高）槓桿的假設。槓桿管理的方式有很多，例如海龜規則中利用帳戶單位和價值量波動性來確定權重和部位。具體來說，先計算每個品種的ATR[2]，衡量其振動幅度；接著，用ATR 乘以合約單位，得到價值量波動性，衡量1 手期貨在價格變化ATR 下的價值總額；最後，用帳戶金額的1% 作為一個單位，除以價值量波動，就得到目標開倉手數，這樣就能剛好保證當價格波動1 個ATR 時，帳戶損失或者盈利保持在帳戶金額的1% 左右。在控制帳戶部位時，要求單一品種不超過4 個單位，高度相關的品種最大不超過6 個單位。海龜法則最終部位取決於交易多少個品種，以及每個品種的單位數，具有一定的彈性。

另外一種常見的槓桿管理策略是目標波動策略（Target Volatility），也叫目標風險策略，在風險控制中比較常見，即透過設定一個目標風險，根據預測波動率來確定部位或槓桿，從而將組合風險控制在該目標。目標波動策略可以用公式表示為：  

$$
L = \frac{\sigma_{tgt}}{\sigma_{t}^{p}}
$$  
其中$\sigma_{tgt}$ 為目標波動水平，$\sigma_{t}^{p}$為組合$p$ 在$t$時刻的波動率估計，$L$為槓桿水平。透過槓桿$L$，可以將目前組合的風險調整到目標水準。例如，如果目標波動率為10%，此時組合波動率為5%，那麼槓桿水準要設為2 倍；如果組合波動率為20%，那麼槓桿為0.5，即50% 的部位水準。  
實證方面，Perchet et al. (2016) 從資產收益率分佈特徵出發，討論目標波動模型有效的邏輯，並應用到了多個資產上。結果發現，目標波動策略在新興市場股票、羅素1000、標普商品指數和美國高收益債券上表現較好，能大幅提高夏普水準。Papageorgiou, Reeves, and Sherris (2017) 使用在美國和澳洲指數數據，發現將目標波動率設為歷史平均波動率時，波動率擇時平均每年能提升收益200bp 左右，並且最大回撤能大幅減少。另外，從指數公司相關產品也能看出目標波動模型的流行。例如，標普道瓊斯編製的目標風險指數序列，中證指數公司編製的各種波控指數。  
對於趨勢追蹤組合來說，Moskowitz, Ooi and Pedersen (2012) 使用單品種目標波動40% 的做法，由於沒有考慮品種之間動態的相關關係，使得組合實現波動率具有較大的不確定性。因此，最好從組合本身層面使用目標波動水平，才能較好透過槓桿控制組合風險，Baltas (2015) 對此有詳細討論。假設趨勢追蹤組合波動率為$\sigma^{TF}$ ，那麼趨勢追蹤策略可以寫為：
$$
r_{t,t+1}^{CVTF} = \frac{\sigma_{tgt}}{\sigma_{t}^{TF}}\sum_{s=1}^{S_t}w_{t}^{s}r_{t,t+1}^{s}
$$
其中$r_{t,t+1}$為單品種趨勢追蹤收益率。  
以風險平價作為資金分配方式，當目標波動率分別為5%、10%、15%、20% 和25% 時的組合結果如表1。首先，隨著目標風險水準的上升，組合收益也單調性地提高，獲得收益的方式就是提高風險承受水準。其次，實現波動率往往高於目標波動率，主要是由於變異數—協方差估計誤差所導致的。最後看看夏普比率，所有目標風險等級夏普相同，這和前面的分析一致，在保持槓桿水平不變的前提下，可以透過槓桿實現不同的風險偏好。綜上，透過目標波動率模型，能巧妙動態地對趨勢追蹤組合的槓桿進行調節，以實現不同的風險效益要求。

**<font  size=5> 5. 趨勢強弱的度量 </font>**  
一旦確定了計算趨勢的時間尺度，另一個面對的問題是選擇什麼方式測量趨勢的強弱，例如是使用價格突破還是收益率平均值？是使用簡單的移動平均還是對近期資料權重較高的指數平均？這是本節要討論的問題。關於不同trend measures 的比較，Levine and Pedersen (2016) 有過精彩的討論，結論是－－ 先潑盆冷水吧－－ 並沒有哪種度量方式吊打其他的。  
在技​​術層面，各種捕捉趨勢的方法也是層出不窮：基於收益率的時序動量、來自技術分析的均線交叉或通道突破、各種state space 模型例如Kalman 濾波、基於價格的線性迴歸、甚至是頻域分析。當某種方法的回測結果不是那麼給力的時候，人們的第一個反應總是尋找下一個更複雜的方法。然而究其核心，上述方法都是基於價格或收益率的時間序列計算出來的。Levine and Pedersen (2016) 透過使用trend signature plots（趨勢簽名圖）這個概念比較了不同的trend measures 方法，認為不同的方法並沒有本質差異。  
以時序收益率均值，技術分析中的長、短均線交叉，Kalman 濾波，以及價格對時間的時序回歸這些常見的trend measures 為例，它們的trend signature plots 如圖7 所示。在給定的計算趨勢的時間尺度下，這些方法的trend signature plots 顯然沒有它們的名字聽起來差異那麼大。它們雖然形態有些差異，但其本質仍是收益率的某種加權平均。
![](attachments/f10.png)  
除了從數學上證明不同trend measures 並無實質差異外，Levine and Pedersen (2016) 也透過實證進行了說明。它們採用Moskowitz, Ooi, and Pedersen (2012) 一文中的58 種資產為交易對象，在不同時間尺度上分別使用兩種不同的trend measures 來建構策略、考察趨勢策略的效果。得到的結論和本文第 節完全一致：

1. 計算趨勢的時間尺度至關重要；
2. 在給定的時間尺度下，具體使用哪一種trend measure 差異並不大。  

既然不同trend measures 並無本質差異，以下我們就以最簡單的移動平均為例來探討一下趨勢訊號背後的本質。

**<font  size=5> 6. 移動平均的本質 </font>**  
在人們的認知中，趨勢追蹤是一個右側交易策略，它使用某個trend measure 來描述過去一段時間上漲或下跌趨勢的強弱，並假設該趨勢會持續。而無論採取何種trend measure，我們都無法提前預判趨勢何時會結束，因此趨勢策略往往在趨勢發生逆轉時錄得虧損，這也正是趨勢策略的風險收益特徵。在本小節，我們試圖回答的問題是，趨勢策略為什麼注定是一個右側交易策略，為什麼無法預知趨勢什麼時候會結束，即趨勢訊號有明顯的滯後性。這就要從趨勢訊號（以移動平均為例）的本質說起。  
從數學上說，移動平均的本質是一種低通濾波。它的目的是過濾掉時間序列中的高頻擾動，保留有用的低頻趨勢。如何從時間序列中抽取出真正的低頻趨勢呢？以下將從數學上解釋移動平均的本質，並揭示趨勢訊號滯後性的成因。假設我們有一個時間序列$y = \{..., y_{t-2}, y_{t-1}, y_{t}, y_{t+2},...\}$如下圖所示。另外，假設我們有一個作用在時域$t$上的濾波函數$F$（$F$的具體形式由選擇的移動平均演算法決定）。  
![](attachments/f11.png)  
在理論上，在任意$t$時刻的低頻濾波（用$x_t$表示）在數學上可以表示為該時間序列$y$和過濾函數$F$在整個時域上的捲積，即  
$$
x_t = \sum_{i=- \infty }^{\infty}F_i y_{t-i}
$$  
其中，$F_i$為濾波函數$F$在時刻$i$的取值（可以理解為$y_{t − i}$的權重；例如簡單平均就是等權）。在實際中通常使用長度為$T$的視窗來代替上面這個無窮級數，這意味著過濾函數$F$只在視窗長度$T$內有效、在視窗之外為0，如下圖所示：  

![](attachments/f12.png)  

加入長度為$T$的視窗函數後，在時刻$t的低頻濾波變為該時間序列$y$和過濾函數$F$在這個視窗內的捲積：
$$
x_t = \sum_{i=t-(T-1)/2 }^{t+(T-1)/2}F_i y_{t-i}
$$  
上述公式最大的問題是：在計算$t$時刻的低頻分量時，就利用了未來的數據。換句話說，理論上的低通濾波（或移動平滑）必須用$t$時刻之後的數據，其假設所有數據都發生後再在全局計算所有時點的低頻分量。但這在即時數據中是不可能的，因為在任何當前時刻$t$，我們都沒有未來數據可以利用。由於現實中無法使用$t−( T−1 ) / 2$到$t+( T−1 ) / 2$之間的數據，只能退而求其次將整個視窗在時間軸上向左平移$( T−1 ) / 2$個單位、使用$t−( T−1 )$到$t$之間的數據來計算$x_t$，如下圖 所示。
![](attachments/f13.png)  
如此處理後，對於即時數據，在當下時刻$t$的低頻濾波變為該時間序列$y$和過濾函數$F$在$t−T+1$到$t$之間的捲積：  
$$
x_t = \sum_{i=t-T+1 }^{t}F_i y_{t-i}
$$  
以簡單移動平均為例，在視窗$T$內，濾波函數在每個時點的取值都是$1 / T$（等權）。利用上述公式計算得到的其實是$t−( T−1 ) / 2$時刻、而非$t$時刻的低頻趨勢，而我們卻只能把它當作$t$時刻的低頻趨勢使用，如此就產生了
$( T−1 ) / 2$的滯後。**沒有未來數據便是趨勢訊號滯後的根本原因**。  

從上述公式可知，計算均線的時間窗口$T$越大，滯後$( T−1 ) / 2$也就越多（例如我們看股票數據裡面MA20、MA30、MA50 等日均線，隨著均線窗口$T$增大，得到的移動平滑曲線也越滯後）。正是由於滯後性和計算均線計算均線視窗的關係，人們本能上更抵觸使用長時間視窗來計算趨勢。同時，人們也更偏好使用指數加權或其他複雜的加權方法，以期更快捕捉趨勢變化的原因（業界為此開發出了很多不同的移動平均演算法），但沒有未來數據造成的滯後性使得趨勢跟踪注定成為右側交易策略。

**<font  size=5> 7. 趨勢追蹤策略的特點 </font>**  
**7.1 危機阿爾法**  
以趨勢追蹤為主要策略的CTA 基金，由於在2008 年金融危機中出色的收益，一舉成名成為了人人皆知的回撤保護器。在相關性分析時，我們發現趨勢追蹤策略和A 股指數相關性為-0.09，起到了良好的分散化作用。實際上不止如此，除了能為現有的資產類型帶來多樣性外，它在市場大跌時的表現更令人津津樂道，因此常被稱為**危機阿爾法（crisis alpha）**。  

Kaminski (2011) 將危機阿爾法定義為：Profits which are gained by exploiting the persistent trends that occur across markets during times of crisis.

當市場因恐慌性下跌時，大部分資產都被拋售，市場情緒極度低迷，而趨勢追蹤可以透過做空在市場下跌時獲得收益。Hamill, Rattray, and Van Hemert (2016) 詳細研究了趨勢追蹤在股票和債券大跌時的表現。將股票指數與公債指數依季度殖利率分別分為5 組，統計每組同期趨勢追蹤策略的表現，結果如圖11 所示。在股票和債券表現最糟糕的一組，趨勢追蹤反而能獲得較高的收益，明顯超過其他組別以及平均值，組合中加入趨勢追蹤能有效燙平波動。  

![](attachments/f14.png)  

Hurst, Ooi, and Pedersen (2017) 列出了美國股市10 大危機時趨勢追蹤的表現（圖12），包括2008 年全球金融危機、網路泡沫和1987 年大跌等。可以看到，在每次股市大調整時，趨勢追蹤都能特立獨行。10 次股市危機中，趨勢追蹤有8 次均能錄得正收益，兩次負收益均較小，和股市相比簡直不值一提；如果將80% 的資金分配給60/40 組合，剩下下的20% 資金分配給趨勢跟踪，能夠有效提升夏普比率並控制回撤。  

![](attachments/f15.png)  

回到台灣市場，趨勢追蹤是否也具有危機阿爾法的功效？為了回答這個問題，圖13 展示了台股指數月收益率相對於趨勢跟踪的散點圖，可以看到，趨勢跟踪呈現出Moskowitz, Ooi and Pedersen (2012)所說的趨勢追蹤微笑形態（time-series momentum smile），當市場大幅下跌或大幅上漲時，均能獲得不錯的收益，體現出大漲時能錦上添花，大跌時也能雪中送碳的優良品質。尤其是在市場恐慌性下跌的時候，趨勢追蹤由於一路賣空，能起到很好的對沖作用。  

![](attachments/f16.png)  


為了進一步探討趨勢追蹤的回撤保護功能，下圖選擇了5 次台股回撤，包括2000股市大崩盤、2008 年國際金融危機、2015 年股市大崩盤以及2018、2022年漫漫陰跌。可以看到，5 次大回撤中，趨勢追蹤有3 次能獲得正收益，尤其是2008 年時，台股腰斬了超過60%，趨勢追蹤旗開得勝獲得了超過50% 的收益；即使兩次收益為負，也只是皮外傷，和同期的台股相比要小得多。

**7.2 多空不對稱性**  
熟悉因子投資的同學應該深有體悟，空頭端貢獻對多空組合的貢獻往往要大，尤其是對量價類因子。在趨勢追蹤策略裡，也存在類似的多空收益不對稱性，不同的是前者是在橫斷面的多空組合，後者是在時間序列上的多空訊號。對每一個品種，將趨勢追蹤拆分成只做多（Long-Only）和只做空（Short-Only）兩個子策略。下圖展示了每個品種兩個子策略的表現，其中左軸表示子策略的年化收益，右軸表示單一品種買入持有的年化收益。可以看到，越靠近左軸的品種，年化收益越低，長期走勢向下，做空比做多貢獻更大；越靠近右軸的品種，年化收益越高，長期走勢向上，做多比做空貢獻較大。不同預期收益率的品種，趨勢追蹤策略多頭和空頭貢獻度的差異，即為多空收益不對稱性。

![](attachments/f18.png)  

Zakamulin (2016) 在討論標普500 指數的擇時策略時也討論過這個特徵。當移動平均線出現死叉時，可以有兩種做法，要么賣出開空，要么空倉持有現金，To Short or Not to Short？如果死叉訊號夠可靠，前者能獲得更高的空頭端收益，賣空當然是更優的選擇。然而，兩種方式的回測結果表明，空倉持有現金策略要比賣出開空擁有更高的夏普，後者可能會增厚投資收益，但是是以犧牲波動率為代價的，並不划算。Jordà et al. (2019) 統計過短期國債、長期國債、股票和房地產四類資產在多個國家近150 年的收益數據，發現美國股票市場長期實際收益8.26%，全球範圍加權股票市場實際收益7.89 %，相當可觀。對於股票這類長期上漲的資產，要嘛做多要嘛空倉，做空可能不划算；換句話說，盡量不要與市場貝塔為敵。


**7.3 勝率和盈虧比**  

Grinold (1989) 提出一個主動投資管理的經驗定律，即策略的資訊比率取決於兩個面向：資訊係數和預測廣度，用公式表達即為：  
$$
IR = IC \times \sqrt{N}
$$  
此定律理解起來也很直觀，資訊係數IC 代表策略的資訊含量和預測能力，投資廣度代表獨立預測的數量。前者為投資的深度，後者為投資的廣度，最終結果取決於兩個角度的結合。趨勢追蹤策略也有類似的概念：盈虧比、勝率和交易次數，其中盈虧比和勝率決定了趨勢追蹤的品質。下圖統計了單品種趨勢追蹤策略的勝率和損益比。從圖中可以看到，趨勢追蹤策略勝率平均略高於50%；再來看看盈虧比，絕大部分品種的盈虧比都在1.0 以上，保證了趨勢追蹤長期的獲利性。

**<font  size=5> 9. 失效了嗎 ? </font>**

**9.1 容量**  
2008 年的一戰成名，讓趨勢追蹤暴露在聚光燈下，蜂擁而至的關注帶來了資金的湧入。Baltas and Kosowski (2013) 對此進行了介紹。2011 年底在所有兩萬億美元對沖基金中，CTA 已經超過了3000 億，CTA 表現不盡如意，一個可能的因素是容量達到了天花板。
Baltas and Kosowski (2013) 使用了兩種方法來探討市場容量是否會影響趨勢追蹤策略。第一個方法進行迴歸分析，以申購資金流入為自變量，趨勢追蹤表現為因變數進行迴歸。結果表明，回歸係數為負但不顯著，表明資金流入並不會影響趨勢追蹤模型的表現。第二個方法進行容量測算，假設系統CTA（systematic CTA）產業的整個AUM 都投資於趨勢追蹤策略，結果發現大多數品種所需的部位並沒有超過該品種的總持倉量；實際上，趨勢追蹤模型的部位只佔整體衍生性商品市場的一小部分。因此，趨勢追蹤表現不好，和容量關係不大。  

**9.2 資產相關性增加**  
Georgopoulou and Wang (2016) 也對趨勢追蹤近年來較差的表現進行了反思，認為趨勢模式其實並沒有發現根本性的改變，主要原因要歸咎於資產之間的相關性發生了變化。具體來說，金融危機後，各國中央銀行普遍實行量化寬鬆的貨幣政策，以刺激經濟的發展，導致各類資產之間的相關性關係遭到了破壞和扭曲。和截面動量不同，時序動量策略喜歡相關性低的資產。而不斷增加的相關性導致獨立的趨勢越來越少，在一定程度上削弱了趨勢追蹤的效果。為了證實這個猜想，Georgopoulou and Wang (2016) 以趨勢追蹤收益率為因變量，以量化寬鬆啞變量QE 為自變量進行回歸，結果表明QE 係數顯著為負，表明量化寬鬆確實會影響趨勢效果。

另外，Baltas (2015) 和Baltas and Kosowski (2017) 同時指出，資產之間相關性的增加可能是趨勢追蹤表現較差的原因之一。Baltas (2015) 利用月內資料計算平均相關係數，接著依相關係大小分為4 組，統計每組趨勢追蹤的表現。結果表明，相關係數越高，收益越低。Baltas and Kosowski (2017) 也發現，考慮相關係數有助於改善趨勢追蹤策略的表現。  

**9.3 沒有趨勢**  
AQR 在其2019 年的報告Babu et al. (2019) 中，也詳細分析了趨勢追蹤折戟的根源。首先，金融危機後，較低的無風險報酬率，可能是造成趨勢追蹤表現較差的因素之一（但不是主要原因）。這裡理解起來很容易，因為期貨是保證金交易，無需全部佔用資金，現金管理能增厚趨勢追蹤的收益，較低的無風險收益自然增厚較少。其次，AQR 認為缺乏分散性並不是趨勢追蹤的原因。為了確認這個問題，AQR 定義了分散乘數（Diversification Multiplier），計算方式為趨勢追蹤組合夏普比率除以單品種趨勢追蹤夏普比率均值，這個值越大說明分散效果越好。透過比較全樣本（1880 ~ 2018年）和2010 到2018年的分散乘數，發現兩者差異幾乎沒啥差別，顯示分散性不足不應背鍋。最後將矛頭指向了趨勢缺乏。如圖22，橫軸為資產的平均漲跌幅度，代表趨勢大小，縱軸為趨勢追蹤策略的表現。可以看到，兩者呈現出明顯的正相關關係，資產趨勢越強勁，趨勢追蹤收益也越漂亮。然而在2010 到2018 年，資產趨勢總體不大，是趨勢追蹤表現差勁的罪魁禍首。話雖如此，AQR 對趨勢追蹤的表現並不悲觀。如果日後資產趨勢幅度增大，趨勢追蹤依然會大有可為。

**<font  size=5> 10. 質疑之聲 </font>**  

雖然趨勢追蹤在很長一段時間內都證明了自己，然而並不是所有人都是它的擁躉。

Kim, Tse, and Wald (2016) 對時間序列動量了進行反思，發現時間序列動量只有使用波動率調整時才會優於買入持有和截面動量策略，沒有進行波動率調整的時間序列動量和買入持有策略沒有顯著差異，進行波動率調整的買入持有策略和進行波動率調整的時間序列動量也沒有明顯差異。因此得出結論，時間序列動量之所以有效，是因為使用了槓桿。

除此之外，Huang et al. (2020) 也更是以Time-series momentum: Is it there? 這樣頗有火藥味的題目在金融學頂刊發文抨擊了Moskowitz, Ooi, and Pedersen (2012)。要知道，這可是本文實證所依賴的基礎，這就很有必要拿出來說道說道。

Huang et al. (2020) 使用與Moskowitz, Ooi, and Pedersen (2012) 相同的資產進行了研究。他們首先用檢驗了不同資產未調整的收益對未來收益的預測能力，並發現只有8（3）種資產在10%（5%）的顯著水準下有顯著的預測能力。樣本外表現則更加糟糕，有45 種資產的樣本外R-squared 是負的。他們接著複製了Moskowitz, Ooi, and Pedersen (2012) 的面板回歸，雖然得到了大體相同的結果，但卻指出Moskowitz, Ooi, and Pedersen (2012) 的回歸中不帶截距項，這意味著資產有相同的夏普比率這個重要的隱含假設。但顯然，股票、商品同外匯和債券的夏普比率並不相同。

Huang et al. (2020) 據此指出，需要在模型中加入不同資產的平均（波動率調整後）收益以控制固定效應的影響，否則估計結果會有向上的偏差。該文用了兩種bootstrap 方法來進行調整，並發現控制了固定效應後，滯後的單月收益普遍都沒有顯著的預測能力，而過去一段時間的累計收益雖仍有顯著的預測力，但顯著性也大幅下滑了。Huang et al. (2020) 進一步檢驗了樣本外表現，發現過去一段時間的收益在股票市場有顯著的樣本外預測能力，但在其他市場則沒有，尤其是債券和外匯市場上，樣本外R- squared 普遍為負。

整體來看，Huang et al. (2020) 認為過去一段時間的收益沒有顯著的預測力，因而Moskowitz, Ooi, and Pedersen (2012) 提出的時序動量演算法基礎並不牢固。他們進而建構了一個不需要收益率預測的TSH 策略，即根據資產歷史平均收益符號來決定交易方向。檢定結果表明，等權重的時序動量和TSH 策略並沒有顯著的不同，因此得出結論，時序動量可能是由於品種之間歷史收益不同引起的。換言之，時序動量的超額報酬更多來自於持倉獲得的風險溢酬。

本節把不一樣的聲音呈現給位讀者，希望以此能引起各位小夥伴的獨立思考。

**<font  size=5> 11. 結語 </font>**  
趨勢追蹤廣泛存在與各大類資產之中，且其成因具有充分的邏輯。從技術層面來看，一個成功的趨勢策略，最重要的是計算趨勢的時間尺度，策略的參數應適應資產的報酬率特性。當確定了時間尺度後，具體使用哪個trend measure 並無本質區別，把更多的精力放在倉位管理來控制風險可能是更好的選擇。

從風險角度來看，趨勢策略屬於divergent risk taking 策略。它雖然不是聖杯，但它的收益率分佈滿足聖杯分佈，這使得它具備危機阿爾法等優點，極具配置價值。不過值得一提的是，有研究表明，過度使用槓桿對這類策略的收益和風險的放大是非線性的。因此，科學、有效的風控是使用趨勢策略的重要一環。

雖然趨勢追蹤策略的收益率分佈呈現正偏態，但從心理層面來看，想要成功使用趨勢追蹤策略就必須要對它的風險收益特徵有清醒的認識。單一資產上的趨勢策略的夏普比率通常不高，但把不同低相關資產的趨勢策略組合在一起能顯著提升portfolio 的夏普比率。即便如此，趨勢策略在短到中期內（例如1 年）也非常有可能出現持續的回檔。駕馭趨勢策略不在於理解它長期來看的正期望，而是在於理解趨勢追蹤策略的風險收益特徵，並欣然接受它短期內潛在較差的表現。

作為一個古老但頗俱生命力的策略，趨勢追蹤策略值得我們認真學習和研究。
