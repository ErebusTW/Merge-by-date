```python
query = """SELECT  [日期]
      ,[股票代號]
      ,[股票名稱]
      ,[資使用率]
      ,[融資維持率(%)]
  FROM [X01].[dbo].[日融資券排行] WITH(NOLOCK) where 日期 >= 20140506 order by 日期 """  

#FROM後填入表格路徑  WHERE指令可以選定只讀取部分數據的條件 order by 可按照特定行排序 若要使用降序 可以寫成 order by 日期 desc
#延伸閱讀可Google "Sql語法" ，也可以問我

df_Financing = pd.read_sql(query, db)
```

```python
df_Financing
```

**<font  size=6>
針對融資維持率進行樞紐操作
</font>**

```python
df_pivot = df_Financing.pivot('日期','股票代號','融資維持率(%)')#可以把融資維持率改為數據集裡的其他columns
```

```python
#pivot後日期會變為索引
df_pivot
```

```python
#若要日期也列為一個column,可以reset_index
df_pivot.reset_index(drop = False)
```

**<font  size=6>
逆樞紐
</font>**

```python
#若要將樞紐後的數據還原，可以用melt指令
#注意，要使用melt前我們要reset_index把日期從索引轉乘columns
#逆樞紐後的值column名稱為value，我們可以rename回來
df_melt = pd.melt(df_pivot.reset_index(drop=False),['日期']).rename(columns={'value':'融資維持率(%)'})
```
